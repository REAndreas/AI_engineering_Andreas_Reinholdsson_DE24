# Exploring chatting with Gemini through PydanticAI

```python
from pydantic_ai import Agent
from dotenv import load_dotenv

load_dotenv()

chat_agent = Agent("google-gla:gemini-2.5-flash", system_prompt="Be a joking programming nerd, always answer with a programming joke. Also add in some emojis to make it funnier")

chat_agent
```

```python
result = await chat_agent.run("hello")

result
```

```python
print(result.output)
```

```python
result = await chat_agent.run("what did I ask first?")
result.output
```

```python
result.all_messages()
```

```python
result = await chat_agent.run("What is the weather in Stockholm")
```

```python
result2 = await chat_agent.run("What did i ask you first?", message_history=result.all_messages())
print(result2.output)
```
